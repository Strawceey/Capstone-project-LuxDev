```python
import pandas as pd
```

```python
#Load the dataset
data = pd.read_csv("tmdb_5000_movies.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
data.shape
```

```python
# Check the data types and missing values
data.info()

# Summary statistics for numerical columns
data.describe()

# Check for missing values
data.isnull().sum()
```

```python
# Drop rows with any missing values
data_cleaned = data.dropna()

# Display the first few rows of the cleaned dataset
print(data_cleaned.head())

# Check if there are any missing values left
print(data_cleaned.isnull().sum())
```

```python
import ast

# Convert genres column from string representation of list to actual list
data['genres'] = data['genres'].apply(lambda x: [i['name'] for i in ast.literal_eval(x)])

# Flatten the list of genres and create a new DataFrame
genre_list = data.explode('genres')

# Count the number of movies in each genre
genre_count = genre_list['genres'].value_counts()
print(genre_count)

# Plotting the most popular genres
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
genre_count.plot(kind='bar')
plt.title('Number of Movies by Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.show()
```

```python
# Extract year from release_date
data['release_year'] = pd.to_datetime(data['release_date']).dt.year

# Flatten the list of genres and create a new DataFrame
genre_list = data.explode('genres')

# Group by release year and genre, then count
genre_trends = genre_list.groupby(['release_year', 'genres']).size().reset_index(name='counts')

# Plot genre trends over time
import seaborn as sns

plt.figure(figsize=(14, 8))
sns.lineplot(data=genre_trends, x='release_year', y='counts', hue='genres')
plt.title('Genre Trends Over Time')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.show()
```

```python
# Plotting the distribution of movie ratings
plt.figure(figsize=(10, 6))
sns.histplot(data['vote_average'], bins=20)
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Movies')
plt.show()
```

```python
# Count the number of movies per production company
production_company_count = data['production_companies'].value_counts().head(10)
print(production_company_count)

# Average rating per production company
production_company_rating = data.groupby('production_companies')['vote_average'].mean().sort_values(ascending=False).head(10)
print(production_company_rating)

# Plotting top production companies by movie count
plt.figure(figsize=(10, 6))
production_company_count.plot(kind='bar')
plt.title('Top 10 Production Companies by Number of Movies')
plt.xlabel('Production Company')
plt.ylabel('Number of Movies')
plt.show()

# Plotting top production companies by average rating
plt.figure(figsize=(10, 6))
production_company_rating.plot(kind='bar')
plt.title('Top 10 Production Companies by Average Rating')
plt.xlabel('Production Company')
plt.ylabel('Average Rating')
plt.show()
```

```python
# Number of movies released per year
movies_per_year = data['release_year'].value_counts().sort_index()

plt.figure(figsize=(14, 8))
movies_per_year.plot(kind='line')
plt.title('Number of Movies Released Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.show()

# Box office trends if box office data is available
if 'revenue' in data.columns:
    plt.figure(figsize=(14, 8))
    data.groupby('release_year')['revenue'].sum().plot(kind='line')
    plt.title('Box Office Trends Over the Years')
    plt.xlabel('Year')
    plt.ylabel('Total Revenue')
    plt.show()
```
